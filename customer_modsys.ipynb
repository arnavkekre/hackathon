## NEW AND IMP PORTION USING HSTACK FOR STACKING TFDIF'D DATA WITH OTHER FEATURES

```python
from scipy.sparse import hstack
```

```python
X_tfidf = tfidf.fit_transform(X['transformed_tweet'])
```

**# Other features (convert booleans to int so the model can handle them)**

```python
other_features=X.drop(columns=['transformed_tweet']).astype(int)
```

**# Convert other_features to sparse (so dimensions match TF-IDF)**

```python
from scipy import sparse
```

```python
other_sparse = sparse.csr_matrix(other_features.values)
```

**FINAL**

```python
X_final = hstack([X_tfidf, other_sparse])
```

```python
X_final
```

```python
from sklearn.metrics import classification_report
```

```python
cnb.fit(X_final, y_encoded)
```

```python
# Get user input
new_tweet = input("Enter your post: ")

# Transform the tweet
transformed_new_tweet = transform_tweet(new_tweet)

# Vectorize the transformed tweet using the fitted tfidf vectorizer
new_tweet_tfidf = tfidf.transform([transformed_new_tweet]) # tfidf.transform expects an iterable

# Create a sparse matrix for the other features with the correct number of columns (95 columns in this case)
# Assuming all 'other features' are absent for a single input tweet, we can create a matrix of zeros
import numpy as np
from scipy.sparse import csr_matrix

# The number of 'other features' columns is the total number of features in X_final minus the number of TF-IDF features
num_other_features = X_final.shape[1] - new_tweet_tfidf.shape[1]
other_features_sparse = csr_matrix(np.zeros((1, num_other_features)))


# Horizontally stack the TF-IDF features and the other features
new_tweet_final = hstack([new_tweet_tfidf, other_features_sparse])


# Make a prediction using the trained model
prediction = cnb.predict(new_tweet_final)

# Decode the prediction back to the original sentiment label
predicted_sentiment = lb.inverse_transform(prediction)

print(f"The predicted sentiment for your post is: {predicted_sentiment[0]}")
```

```python
print(classification_report(y_encoded, cnb.predict(X_final)))
```
